```r
# Rejection sampling function for points within a unit disk in the first quadrant (vectorized)
accept_reject <- function(n) {
  # Generate random points from a uniform square
  x <- runif(n)
  y <- runif(n)

  # Efficiently check if points fall within the unit disk
  accepted <- (x^2 + y^2 <= 1)

  # Return only accepted samples
  return(cbind(x[accepted], y[accepted]))
}

# Generate 1000 samples
samples <- accept_reject(1000)
```

```r
# Plot the samples
plot(samples[, 1], samples[, 2],
      col = "blue",
     main = "Samples within a unit disk",
     xlab = "X", ylab = "Y")

# Add a circle to represent the unit disk
#circle(0, 0, radius = 1, col = "gray", lty = 2)
```
